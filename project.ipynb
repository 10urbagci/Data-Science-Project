```python
#independent features
X = df[['price','price_usd','volume','volume_usd']]
#Dependent features
y = df['base']
     
```

```python
#Data traning
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train,y_train)
```

```python
#Logistic Regression estimation
y_pred1 = logr.predict(X_test)
print('Predict result for Logistic Regression:')
print(y_pred1)
```

```python
#Real test data
y_test
```

```python
print('Confusion Matrix for Logistic Regression')
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred1)
print(cm)
```

```python
print('Logistic Regression Accuracy Score:')
acc = accuracy_score(y_test,y_pred1)
print(acc)
```

When we look at the Logistic Regression estimation results and the real test data, the Logistic Regression algorithm made only one error in the estimation. As a result, it achieved a 94% success rate in the Logistic Regression algorithm. Confusion matrix was used to measure the performance of the classification model. Correct predictions and incorrect predictions appear in this matrix. The same method has been applied in Decision Tree and SVC classification algorithms.Finally, the actual test data and the model predicted results are mixed.The effect of prices and volumes on bitcoin classification has been studied.
